```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visual = pd.read_csv('../Information_Mart/orders_products_details.csv')
```

```python
visual.info()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=visual, x='order_status_desc')
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=visual, x='category_id')
plt.title('Order Quantity by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
late_delivery_counts = visual['late_delivery'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(late_delivery_counts, labels=['On Time', 'Late'], autopct='%1.1f%%', startangle=140)
plt.title('Late Delivery Percentage')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=visual, x='category_id', y='list_price_egp', estimator=pd.np.mean)
plt.title('Bar Plot of Average List Price (EGP) by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average List Price (EGP)')
plt.xticks(rotation=45)
plt.show()
```
