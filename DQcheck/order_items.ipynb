```python
import pandas as pd
```

```python
df_order_items = pd.read_csv('../landing/order_item/order_item.csv')
df_order_items.info()
```

```python
df_order_items.duplicated()
```

```python
df_products = pd.read_csv('../landing/products/products.csv')
df_products.info()
```

```python
df_order_items = df_order_items[df_order_items['product_id'].isin(df_products['product_id'])]
df_order_items
```

```python
df_order_items.isnull().sum()
```

```python
df_order_items = df_order_items[(df_order_items['list_price'] >= 200.0 ) & (df_order_items['list_price'] <= 10000.0 )]
df_order_items
```

```python
df_order_items['item_id'] = pd.to_numeric(df_order_items['item_id'], errors = 'coerce')
```

```python
df_order_items.isnull().sum()
```

```python
df_order_items['item_id'].fillna(method = 'ffill', inplace = True)
```

```python
df_order_items.isnull().sum()
```

```python
df_order_items.duplicated().sum()
```

```python
df_order_items.drop_duplicates()
```

```python
df_order_items.to_csv('../staging/order_items/order_items.csv', index = 0)
```
